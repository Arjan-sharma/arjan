```python
arr = [1, 4, 3, 2, 6, 5]
print(f"Original array: {arr}")

def reverse_array_manual(arr):
    left = 0
    right = len(arr) - 1
    while left < right:

        arr[left], arr[right] = arr[right], arr[left]

        left += 1
        right -= 1

reverse_array_manual(arr)
print(f"Reversed array (manual): {arr}")
```

```python
arr = [1, 4, 3, 2, 6, 5, 0, -1, 10] # Re-initialize the array for demonstration
print(f"Original array: {arr}")

# Method 2: Manual iteration
def find_min_max_manual(arr):
    if not arr: # Handle empty array case
        return None, None

    min_val = arr[0]
    max_val = arr[0]

    for element in arr[1:]:
        if element < min_val:
            min_val = element
        if element > max_val:
            max_val = element
    return min_val, max_val

min_element_manual, max_element_manual = find_min_max_manual(arr)

print(f"Minimum element (manual): {min_element_manual}")
print(f"Maximum element (manual): {max_element_manual}")
```

```python
def find_kth_smallest(arr, k):
    if not arr:
        return None, "Array is empty"
    if k <= 0 or k > len(arr):
        return None, f"k ({k}) is out of bounds for array of size {len(arr)}"

    # Sort the array in ascending order
    arr.sort()

    # Return the (k-1)th element for 0-based indexing
    return arr[k-1], None

arr_new = [12, 3, 5, 7, 19]
k_new = 1
print(f"\nOriginal array: {arr_new}")
print(f"k = {k_new}")
kth_smallest_element_new, error_message_new = find_kth_smallest(arr_new, k_new)

if error_message_new:
    print(f"Error: {error_message_new}")
else:
    print(f"The {k_new}st smallest element is: {kth_smallest_element_new}")

k_out_of_bounds = 6
print(f"\nOriginal array: {arr_new}")
print(f"k = {k_out_of_bounds}")
kth_smallest_element_oob, error_message_oob = find_kth_smallest(arr_new, k_out_of_bounds)

if error_message_oob:
    print(f"Error: {error_message_oob}")
else:
    print(f"The {k_out_of_bounds}th smallest element is: {kth_smallest_element_oob}")
```

```python
def find_union(arr1, arr2):
    # Convert both arrays to sets to get distinct elements
    set1 = set(arr1)
    set2 = set(arr2)

    # Find the union of the two sets
    union_set = set1.union(set2)

    # Convert the resulting set back to a list
    return list(union_set)

# Example 1
a1 = [1, 2, 3, 4, 5]
b1 = [1, 2, 3, 6, 7]
union1 = find_union(a1, b1)
print(f"Array 1: {a1}")
print(f"Array 2: {b1}")
print(f"Union of arrays: {union1}")

print("\n" + "-"*30 + "\n")

# Example 2 with duplicate elements within arrays
a2 = [8, 4, 2, 1, 8, 1]
b2 = [4, 7, 9, 2, 7]
union2 = find_union(a2, b2)
print(f"Array 1: {a2}")
print(f"Array 2: {b2}")
print(f"Union of arrays: {union2}")
```

```python
def find_array_union(arr1, arr2):
    """
    Finds the union of two arrays, returning all distinct elements.
    """
    # Convert both arrays to sets to automatically handle distinct elements
    set1 = set(arr1)
    set2 = set(arr2)

    # Find the union of the two sets
    union_set = set1.union(set2)

    # Convert the resulting set back to a list and return
    return list(union_set)

# Example usage:
array_a = [10, 20, 30, 40, 10]
array_b = [30, 50, 60, 20]

union_result = find_array_union(array_a, array_b)
print(f"Array A: {array_a}")
print(f"Array B: {array_b}")
print(f"Union of Arrays: {union_result}")

print("\n" + "-"*30 + "\n")

# Another example:
array_c = [1, 2, 3, 4]
array_d = [5, 6, 7, 8]
union_result_2 = find_array_union(array_c, array_d)
print(f"Array C: {array_c}")
print(f"Array D: {array_d}")
print(f"Union of Arrays: {union_result_2}")
```

```python
def find_largest_element(arr):
    if not arr:
        return None, "Array is empty"

    max_element = arr[0]
    for element in arr:
        if element > max_element:
            max_element = element
    return max_element, None

# Example usage:
my_array = [1, 5, 2, 8, 3, 9, 4]
largest, error = find_largest_element(my_array)

if error:
    print(f"Error: {error}")
else:
    print(f"Original array: {my_array}")
    print(f"The largest element is: {largest}")

another_array = [100, 20, 300, 50]
largest_2, error_2 = find_largest_element(another_array)

if error_2:
    print(f"Error: {error_2}")
else:
    print(f"\nOriginal array: {another_array}")
    print(f"The largest element is: {largest_2}")

empty_array = []
largest_empty, error_empty = find_largest_element(empty_array)

if error_empty:
    print(f"\nError: {error_empty}")
```

```python
def rotate_array_clockwise_by_one(arr):
    """
    Rotates an array by one position in a clockwise direction.
    The last element moves to the first position, and all other elements shift right.
    """
    if not arr or len(arr) == 1:
        return arr # No rotation needed for empty or single-element arrays

    last_element = arr[-1]
    for i in range(len(arr) - 1, 0, -1):
        arr[i] = arr[i - 1]
    arr[0] = last_element
    return arr

# Example usage:
my_array_1 = [1, 2, 3, 4, 5]
print(f"Original array: {my_array_1}")
rotated_array_1 = rotate_array_clockwise_by_one(my_array_1)
print(f"Rotated array: {rotated_array_1}")

print("\n" + "-"*30 + "\n")

my_array_2 = [10, 20, 30]
print(f"Original array: {my_array_2}")
rotated_array_2 = rotate_array_clockwise_by_one(my_array_2)
print(f"Rotated array: {rotated_array_2}")

print("\n" + "-"*30 + "\n")

my_array_3 = [7]
print(f"Original array: {my_array_3}")
rotated_array_3 = rotate_array_clockwise_by_one(my_array_3)
print(f"Rotated array: {rotated_array_3}")

print("\n" + "-"*30 + "\n")

empty_array = []
print(f"Original array: {empty_array}")
rotated_empty_array = rotate_array_clockwise_by_one(empty_array)
print(f"Rotated array: {rotated_empty_array}")
```

```python
def max_subarray_sum(arr):
    if not arr:
        return 0, "Array is empty, max sum is 0"

    # Initialize max_so_far and current_max
    # These variables will store the maximum sum found globally and for the current subarray, respectively.
    max_so_far = arr[0]
    current_max = arr[0]

    # Iterate through the array starting from the second element
    for i in range(1, len(arr)):
        # For each element, decide whether to extend the current subarray
        # or start a new one. The current_max is the maximum of the current element
        # itself or the current element added to the previous current_max.
        current_max = max(arr[i], current_max + arr[i])

        # Update the overall maximum sum found so far
        max_so_far = max(max_so_far, current_max)

    return max_so_far, None

# Example usage:
array1 = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
max_sum1, error1 = max_subarray_sum(array1)
if error1:
    print(f"Error: {error1}")
else:
    print(f"Original array: {array1}")
    print(f"Maximum subarray sum: {max_sum1}") # Expected output: 6 ([4, -1, 2, 1])

print("\n" + "-"*30 + "\n")

array2 = [1]
max_sum2, error2 = max_subarray_sum(array2)
if error2:
    print(f"Error: {error2}")
else:
    print(f"Original array: {array2}")
    print(f"Maximum subarray sum: {max_sum2}") # Expected output: 1 ([1])

print("\n" + "-"*30 + "\n")

array3 = [5, 4, -1, 7, 8]
max_sum3, error3 = max_subarray_sum(array3)
if error3:
    print(f"Error: {error3}")
else:
    print(f"Original array: {array3}")
    print(f"Maximum subarray sum: {max_sum3}") # Expected output: 23 ([5, 4, -1, 7, 8])

print("\n" + "-"*30 + "\n")

array4 = [-2, -3, -1, -5]
max_sum4, error4 = max_subarray_sum(array4)
if error4:
    print(f"Error: {error4}")
else:
    print(f"Original array: {array4}")
    print(f"Maximum subarray sum: {max_sum4}") # Expected output: -1 ([-1])

print("\n" + "-"*30 + "\n")

empty_array = []
max_sum_empty, error_empty = max_subarray_sum(empty_array)
if error_empty:
    print(f"Error: {error_empty}")
```

```python
def search_insert_position(nums, target):
    """
    Given a sorted array of distinct integers and a target value,
    return the index if the target is found. If not, return the index
    where it would be if it were inserted in order.
    """
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    # If the loop finishes, 'left' points to the insertion position
    return left

# Example usage:

# Target found
nums1 = [1, 3, 5, 6]
target1 = 5
print(f"Array: {nums1}, Target: {target1} -> Index: {search_insert_position(nums1, target1)}") # Expected: 2

# Target not found, insert at the end
nums2 = [1, 3, 5, 6]
target2 = 7
print(f"Array: {nums2}, Target: {target2} -> Index: {search_insert_position(nums2, target2)}") # Expected: 4

# Target not found, insert at the beginning
nums3 = [1, 3, 5, 6]
target3 = 0
print(f"Array: {nums3}, Target: {target3} -> Index: {search_insert_position(nums3, target3)}") # Expected: 0

# Target not found, insert in the middle
nums4 = [1, 3, 5, 6]
target4 = 2
print(f"Array: {nums4}, Target: {target4} -> Index: {search_insert_position(nums4, target4)}") # Expected: 1

# Empty array
nums5 = []
target5 = 5
print(f"Array: {nums5}, Target: {target5} -> Index: {search_insert_position(nums5, target5)}") # Expected: 0

# Single element array - target found
nums6 = [5]
target6 = 5
print(f"Array: {nums6}, Target: {target6} -> Index: {search_insert_position(nums6, target6)}") # Expected: 0

# Single element array - target not found (insert after)
nums7 = [5]
target7 = 6
print(f"Array: {nums7}, Target: {target7} -> Index: {search_insert_position(nums7, target7)}") # Expected: 1

# Single element array - target not found (insert before)
nums8 = [5]
target8 = 4
print(f"Array: {nums8}, Target: {target8} -> Index: {search_insert_position(nums8, target8)}") # Expected: 0
```

```python
def two_sum(nums, target):
    """
    Given an array of integers nums and an integer target,
    return indices of the two numbers such that they add up to target.
    Assumes exactly one solution and cannot use the same element twice.
    """
    num_map = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return [num_map[complement], i]
        num_map[num] = i
    # According to the problem statement, there will always be exactly one solution.
    # If for some reason no solution is found (e.g., input doesn't match constraints),
    # you might raise an exception or return an empty list.
    return []

# Example Usage:
nums1 = [2, 7, 11, 15]
target1 = 9
print(f"Array: {nums1}, Target: {target1} -> Indices: {two_sum(nums1, target1)}") # Expected: [0, 1]

nums2 = [3, 2, 4]
target2 = 6
print(f"Array: {nums2}, Target: {target2} -> Indices: {two_sum(nums2, target2)}") # Expected: [1, 2]

nums3 = [3, 3]
target3 = 6
print(f"Array: {nums3}, Target: {target3} -> Indices: {two_sum(nums3, target3)}") # Expected: [0, 1]

nums4 = [1, 5, 9, 2, 6]
target4 = 7
print(f"Array: {nums4}, Target: {target4} -> Indices: {two_sum(nums4, target4)}") # Expected: [0, 4] (indices of 1 and 6)
```

```python
def min_jumps(arr):
    if not arr or len(arr) <= 1:
        return 0, None # No jumps needed for empty or single-element array

    n = len(arr)
    jumps = 0
    max_reach = 0
    steps_possible = 0

    # Iterate through the array
    for i in range(n - 1):
        # Update max_reach (farthest point we can reach from current position)
        max_reach = max(max_reach, i + arr[i])

        # Decrement steps_possible, as we've used one step to get to the current position
        steps_possible -= 1

        # If we run out of steps_possible from the current jump, we need another jump
        if steps_possible < 0:
            return -1, "Cannot reach the end of the array." # Indicate failure

        if steps_possible == 0:
            # If we are at the last position and still have steps_possible, we don't need to jump again.
            # But if we need to make another jump, increment jumps
            jumps += 1

            # If the current position is beyond or at the max_reach, it means we are stuck
            # and cannot move forward, so the end is unreachable.
            if i >= max_reach:
                return -1, "Cannot reach the end of the array." # Indicate failure

            # Update steps_possible for the new jump from the max_reach point
            steps_possible = max_reach - i


    # If we successfully reach or pass the last index, return the total jumps
    return jumps, None

# Example usage with your provided array:
arr_input = [1, 3, 5, 8, 9, 2, 6, 7, 6, 8, 9]
num_jumps, error = min_jumps(arr_input)

if error:
    print(f"Array: {arr_input}")
    print(f"Error: {error}")
else:
    print(f"Array: {arr_input}")
    print(f"Minimum number of jumps to reach the end: {num_jumps}")


print("\n" + "-"*30 + "\n")

# Another example where it's impossible to reach the end
arr_impossible = [3, 2, 1, 0, 4]
num_jumps_impossible, error_impossible = min_jumps(arr_impossible)

if error_impossible:
    print(f"Array: {arr_impossible}")
    print(f"Error: {error_impossible}")
else:
    print(f"Array: {arr_impossible}")
    print(f"Minimum number of jumps to reach the end: {num_jumps_impossible}")
```
